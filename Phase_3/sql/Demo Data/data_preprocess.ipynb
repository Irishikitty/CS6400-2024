# SQL queries

## for loop auto generate SQL queries 

### define func

```python
def generate_insert_queries(df, table_name, table_column_names):
  """Generates SQL INSERT queries from a Pandas DataFrame.

  Args:
    df: The Pandas DataFrame.
    table_name: The name of the table to insert into.
    table_column_names: A list of df column names to include in the query.

  Returns:
    A list of SQL INSERT queries.
  """

  queries = []
  for index, row in df.iterrows():
    if len(df.columns) == 1:
      # Single-column case
      # values = tuple(row[col] for col in list(df.columns))
      query = f"INSERT INTO {table_name} ({df.columns[0]}) VALUES ('{row[table_column_names[0]]}');"
    else:
      # Multiple-column case
      values = tuple(row[col] for col in list(df.columns))
      query = f"INSERT INTO {table_name} ({', '.join(table_column_names)}) VALUES {values};"
    queries.append(query)
  return queries
```

## Write all SQL INSERT queries into sql

```python
import os

def write_queries_to_file(queries, table_name,  filename="team105_p3_demodata.sql"):
  """Writes a list of SQL queries to a file.

  Args:
    queries: A list of SQL queries.
    filename: The name of the output file.
  """

  with open(filename, 'a') as f:
    f.write('-- Inserting '+ table_name + '\n')
    for query in queries:
      f.write(query + '\n')
    f.write('\n')
```
